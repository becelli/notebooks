# KMeans


```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt

_ = plt.figure(figsize=(16, 9), dpi=54)
```

```python
def kmeans(input, k):
    centroids = np.random.rand(k, 2)
    groups = None

    for _ in range(10):
        groups = [[] for i in range(k)]

        for data in range(len(input)):
            min_distance = np.inf
            center_index = -1

            for c in range(k):
                distance = la.norm(centroids[c] - input[data])
                if distance < min_distance:
                    min_distance = distance
                    center_index = c

            groups[center_index].append(input[data])

        for c in range(k):
            if len(groups[c]) > 0:
                centroids[c] = np.mean(groups[c], axis=0)
    return centroids, groups
```

```python
def generate_area(dots, centres):
    x = np.array([])
    y = np.array([])

    for i in range(centres):
        xn = np.random.rand(dots) / 4 + np.random.rand(1) * 2 / 3
        yn = np.random.rand(dots) / 4 + np.random.rand(1) * 2 / 3
        x = np.hstack((x, xn))
        y = np.hstack((y, yn))
    return np.vstack((x, y)).T
```

```python
def generate_random_points(dots):

    x = np.random.rand(dots)
    y = np.random.rand(dots)
    return np.vstack((x, y)).T
```

```python
# data = generate_random_points(100)
data = generate_area(100, 3)
plt.scatter(data[:, 0], data[:, 1])
```

```python
centroids, groups = kmeans(data, 3)
```

```python
for g in groups:
    if len(g) > 0:
        group = np.array(g)
        plt.scatter(group[:, 0], group[:, 1])
plt.scatter(centroids[:, 0], centroids[:, 1], c="black")
```
